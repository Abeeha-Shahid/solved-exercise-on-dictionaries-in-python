```python
# Input text 
text = "count the frequency of words in given text, count frequency"

# Split the text into words
words = text.split()

# Create an empty dictionary to store word frequencies
word_count = {}

# Loop 
for word in words:
    # If the word is already in the dictionary, increment its count
    if word in word_count:
        word_count[word] += 1
    # If the word is not in the dictionary, add it with a count of 1
    else:
        word_count[word] = 1

# Print the word frequencies
print(word_count)
```

```python
#part d 
orders = [
    {"order_id": 1, "product": "Laptop", "category": "Electronics"},
    {"order_id": 2, "product": "Shirt", "category": "Clothing"},
    {"order_id": 3, "product": "Smartphone", "category": "Electronics"},
    {"order_id": 4, "product": "Jacket", "category": "Clothing"},
    {"order_id": 5, "product": "Tablet", "category": "Electronics"},
    {"order_id": 6, "product": "Jeans", "category": "Clothing"},
]

from collections import Counter

# Extract categories from orders
categories = [order["category"] for order in orders]

# Count the frequency of each category
category_count = Counter(categories)

# Find the most popular category
most_popular_category = category_count.most_common(1)[0]

# Print the result
print(f"The most popular category is '{most_popular_category[0]}' with {most_popular_category[1]} orders.")
```

```python
# Product info: name, quantity, price
inventory = {
    "apple": {"quantity": 50, "price": 1.20},
    "banana": {"quantity": 30, "price": 0.80},
}

# Add new product or update quantity
def add_product(name, quantity, price):
    if name in inventory:
        inventory[name]["quantity"] += quantity
    else:
        inventory[name] = {"quantity": quantity, "price": price}

# Update product quantity
def update_quantity(name, quantity):
    if name in inventory:
        inventory[name]["quantity"] = quantity

# Calculate total inventory value
def total_inventory_value():
    return sum(item["quantity"] * item["price"] for item in inventory.values())

# Low-stock report (threshold 10)
def low_stock_report(threshold=10):
    return [name for name, info in inventory.items() if info["quantity"] <= threshold]

# Main loop for user interaction
while True:
    print("\n1. Add/Update Product\n2. Total Inventory Value\n3. Low Stock Report\n4. Exit")
    choice = input("Choose an option: ")

    if choice == "1":
        name = input("Enter product name: ")
        quantity = int(input("Enter quantity: "))
        price = float(input("Enter price: "))
        add_product(name, quantity, price)
    elif choice == "2":
        print(f"Total Inventory Value: ${total_inventory_value():.2f}")
    elif choice == "3":
        low_items = low_stock_report()
        print("Low Stock Products:", low_items if low_items else "None")
    elif choice == "4":
        break
    else:
        print("Invalid option.")
```

```python
# Dictionary storing student name and grades for subjects
students = {
    "Abeeha": {"math": 85, "science": 90, "english": 88},
    "Rabiya": {"math": 78, "science": 85, "english": 82},
    "Haseeba": {"math": 92, "science": 94, "english": 91},
}

# Calculate average grade for each student
def average_grade(grades):
    return sum(grades.values()) / len(grades)

# Find highest and lowest grades in a subject
def subject_grades(subject):
    highest = None
    lowest = None
    for name, grades in students.items():
        grade = grades[subject]
        if highest is None or grade > highest[1]:
            highest = (name, grade)
        if lowest is None or grade < lowest[1]:
            lowest = (name, grade)
    return highest, lowest

# Sort students by average grade
def sorted_by_average():
    sorted_students = []
    for name, grades in students.items():
        avg = average_grade(grades)
        sorted_students.append((name, avg))
    sorted_students.sort(key=lambda x: x[1], reverse=True)
    return sorted_students

# Main Program
while True:
    print("\n1. Average Grade for Each Student")
    print("2. Find Highest and Lowest Grades in a Subject")
    print("3. Sort Students by Average Grade")
    print("4. Exit")
    
    choice = input("Choose an option (1-4): ")

    if choice == "1":
        for name, grades in students.items():
            avg = average_grade(grades)
            print(f"{name}: Average Grade = {avg:.2f}")
    elif choice == "2":
        subject = input("Enter subject (math, science, english): ")
        if subject in next(iter(students.values())):
            highest, lowest = subject_grades(subject)
            print(f"Highest grade in {subject}: {highest[0]} with {highest[1]}")
            print(f"Lowest grade in {subject}: {lowest[0]} with {lowest[1]}")
        else:
            print("Invalid subject.")
    elif choice == "3":
        sorted_students = sorted_by_average()
        print("Students sorted by average grade:")
        for name, avg in sorted_students:
            print(f"{name}: Average Grade = {avg:.2f}")
    elif choice == "4":
        break
    else:
        print("Invalid choice, try again.")
```

```python
#Student gradebook
# Product info: name, price, quantity
products = {"apple": [1.20, 50], "banana": [0.80, 30], "orange": [1.50, 20]}
cart = {}

def add_to_cart(p, q):
    if p in products and products[p][1] >= q: cart[p] = cart.get(p, 0) + q; products[p][1] -= q
    else: print("Not enough stock.")

def remove_from_cart(p, q):
    if p in cart and cart[p] >= q: cart[p] -= q; products[p][1] += q; (cart.pop(p) if cart[p] == 0 else None)
    else: print("Invalid or insufficient quantity.")

def total_cost():
    t = sum(products[p][0] * q for p, q in cart.items())
    return t * 0.9 * 1.08  # 10% discount + 8% tax

while True:
    choice = input("\n1.Add 2.Remove 3.Checkout 4.Exit: ")
    if choice == "1": add_to_cart(input("Product: "), int(input("Quantity: ")))
    elif choice == "2": remove_from_cart(input("Product: "), int(input("Quantity: ")))
    elif choice == "3": print(f"Total: ${total_cost():.2f}"); break
    elif choice == "4": break
```
